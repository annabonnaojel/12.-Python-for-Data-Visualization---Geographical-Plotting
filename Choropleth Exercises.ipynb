```python
init_notebook_mode(connected=True)
```

```python
df=pd.read_csv('country_wise_latest.csv')
```

```python
df.head()
```

```python
import plotly.graph_objects as go

data = go.Choropleth(
    locations=df['Country/Region'],
    locationmode='country names',
    colorscale='viridis',
    reversescale=False,
    text=df['Country/Region'],
    z=df['Confirmed'],
    colorbar={'title': 'Confirmed'}
)
```

```python
layout = go.Layout(
    title='Country wise Latest',
    geo=dict(
        projection=dict(type='mercator')
    )
)
```

```python
choromap = go.Figure(data = [data],layout=layout)
```

```python
iplot(choromap)

choromap.show()
```

```python
df=pd.read_csv('county_facts.csv')
```

```python
df.head()
```

```python
import plotly.graph_objects as go

data = go.Choropleth(
    locations=df['state_abbreviation'],  # or whatever correct column
    locationmode='USA-states',
    colorscale='viridis',
    reversescale=False,
    text=df['area_name'],
    z=df['pop010210'],  # make sure this column name is correct
    colorbar={'title': 'Population'}
)
```

```python
layout = dict(title='US County Level Presidential Results 2012 and 2016',
        geo=dict(
        scope='usa',showlakes=True,
        lakecolor='rgb(85,173,240)')
)
```

```python
choromap = go.Figure(data = [data],layout=layout)
iplot(choromap,validate=False)
```
